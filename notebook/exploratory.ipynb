```python
df['month+year'] = df['MoSold'].astype(str) + '.' + df['YrSold'].astype(str)
df['totalbath'] = df['BsmtFullBath'] + (df['BsmtHalfBath'] * 0.6) + df['FullBath'] + (df['HalfBath'] * 0.5)
```

```python
df.drop(['BsmtFullBath'], axis=1, inplace=True)
df.drop(['BsmtHalfBath'], axis=1, inplace=True)
df.drop(['FullBath'], axis=1, inplace=True)
df.drop(['HalfBath'], axis=1, inplace=True)
df.drop(['MoSold'], axis=1, inplace=True)
```

```python
df.drop(['YrSold'], axis=1, inplace=True)
```

```python
df.drop(['Id','MiscFeature', 'Alley', 'Fence', 'MasVnrType', 'FireplaceQu', 'LotFrontage'], axis=1, inplace=True)
```

```python
df.columns
```

```python
for col in df.select_dtypes(include='number').columns:
    if col != "SalePrice":
        mask = df[col] > 0
        filtered_x = df.loc[mask, col]
        filtered_y = df.loc[mask, 'SalePrice']
        plt.figure(figsize=(6, 4))
        plt.scatter(df[col], df['SalePrice'])
        corr = df[col].corr(df['SalePrice'])
        plt.xlabel(col)
        plt.ylabel('SalePrice')
        plt.title(f'{col} v.s. SalePrice')
        plt.title(f'{col} vs SalePrice\nCorrelation: {corr:.2f}')
        plt.tight_layout()
        plt.show()

```

```python
df.drop(['MiscVal', 'OverallCond', 'MSSubClass', 'PoolArea', 'ScreenPorch', '3SsnPorch', 'EnclosedPorch', 'KitchenAbvGr'], axis=1, inplace=True)
df.drop(['LowQualFinSF', 'BedroomAbvGr', 'BsmtFinSF2'], axis=1, inplace=True)
```
